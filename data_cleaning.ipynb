```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv('data/rating.csv', encoding= 'utf-8')
ratings.head()
```

```python
ratings = ratings.loc[ratings['rating'] != -1,:]
ratings.head()
```

```python
len(ratings) #7.8 millones de entries (with rating: -1) & 6.33 millones (sin rating: -1)
len(ratings['user_id'].unique()) #73.515 usuarios  &  69600 usuarios (sin rating: -1)
ratings.groupby('user_id')['rating'].mean().mean() #mean rating for anime is 8.22
ratings.groupby('user_id')['rating'].agg(['count', 'mean', 'std'])
```
